```python
import pandas as pd
df1 = pd.read_csv('term-test-1-result.csv')
df2 = pd.read_csv('term-test-2-result.csv')
merged_df = pd.merge(df1, df2, on='Registration Number', suffixes=('_1', '_2'))
merged_df['Name'] = merged_df['Name_1']
merged_df = merged_df[['Registration Number', 'Name', 'TT-1 Marks', 'TT-2 Marks']]
merged_df
```

```python
merged_df['Best Term Test Marks'] = merged_df[['TT-1 Marks', 'TT-2 Marks']].max(axis=1)
merged_df['Average Term Test Marks'] = merged_df[['TT-1 Marks', 'TT-2 Marks']].mean(axis=1)
merged_df
```

```python
merged_df = merged_df.drop(columns=['TT-1 Marks', 'TT-2 Marks'])
merged_df
```

```python
attendance_file = "attendance_and_final_marks.csv"
attendance_df = pd.read_csv(attendance_file)
final_df = pd.merge(merged_df, attendance_df, on='Registration Number')
final_df['Final Marks'] = (final_df['Term Final Marks'] * 0.7) + final_df['Average Term Test Marks'] + final_df['Attendance Marks']

final_df.to_csv('final_student_results.csv', index=False)
final_df
```

```python
final_df.to_csv('final result.csv', index=False)
print("Final result saved to 'final result.csv'")
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import seaborn as sns
final_df = pd.read_csv('final result.csv')
final_marks = final_df[['Final Marks']]
kmeans = KMeans(n_clusters=4, random_state=42)
final_df['Cluster'] = kmeans.fit_predict(final_marks)
plt.figure(figsize=(8, 6))
sns.scatterplot(x=final_df['Registration Number'], y=final_df['Final Marks'], hue=final_df['Cluster'], palette='viridis', s=100)
plt.title("K-Means Clustering of Final Marks")
plt.xlabel("Registration Number")
plt.ylabel("Final Marks")
plt.legend(title='Cluster', loc='upper right')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
